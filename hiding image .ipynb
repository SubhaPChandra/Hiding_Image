```python
import cv2


def char_generator(message):
  for c in message:
    yield ord(c)

def get_image(image_location):
  img = cv2.imread(image_location)
  return img

def gcd(x, y):
  while(y):
    x, y = y, x % y

  return x

def encode_image(image_location, msg):
  img = get_image(image_location)
  msg_gen = char_generator(msg)
  pattern = gcd(len(img), len(img[0]))
  for i in range(len(img)):
    for j in range(len(img[0])):
      if (i+1 * j+1) % pattern == 0:
        try:
          img[i-1][j-1][0] = next(msg_gen)
        except StopIteration:
          img[i-1][j-1][0] = 0
          return img

def decode_image(img_loc):
  img = get_image(img_loc)
  pattern = gcd(len(img), len(img[0]))
  message = ''
  for i in range(len(img)):
    for j in range(len(img[0])):
      if (i-1 * j-1) % pattern == 0:
        if img[i-1][j-1][0] != 0:
          message = message + chr(img[i-1][j-1][0])
        else:
          return message
```

```python
from PIL import Image

def text_to_binary(text):
    binary_message = ''.join(format(ord(char), '08b') for char in text)
    return binary_message

def binary_to_text(binary_message):
    text = ''.join(chr(int(binary_message[i:i+8], 2)) for i in range(0, len(binary_message), 8))
    return text

def hide_message(image_path, secret_message, output_path):
    img = Image.open(image_path)

    binary_message = text_to_binary(secret_message)
    binary_message += '1111111111111110'  # Add a delimiter indicating the end of the message

    data_index = 0
    data_length = len(binary_message)

    for i in range(img.width):
        for j in range(img.height):
            pixel = list(img.getpixel((i, j)))

            for color_channel in range(3):  # RGB channels
                if data_index < data_length:
                    pixel[color_channel] = int(format(pixel[color_channel], '08b')[:-1] + binary_message[data_index], 2)
                    data_index += 1

            img.putpixel((i, j), tuple(pixel))

    img.save(output_path)

def reveal_message(image_path):
    img = Image.open(image_path)
    binary_message = ""

    for i in range(img.width):
        for j in range(img.height):
            pixel = list(img.getpixel((i, j)))

            for color_channel in range(3):
                binary_message += format(pixel[color_channel], '08b')[-1]

    delimiter_index = binary_message.find('1111111111111110')
    binary_message = binary_message[:delimiter_index]

    secret_message = binary_to_text(binary_message)
    return secret_message

if __name__ == "__main__":
    # Example usage
    image_path = "/Users/subhaprasadchandra/Downloads/11-2-vector-png-pic.png"
    secret_message = "Hello, this is a hidden message!"
    output_path = "/Users/subhaprasadchandra/Downloads/11-2-vector-png-pic.png"
    #"/Users/subhaprasadchandra/Downloads/png-transparent-pink-cross-stroke-ink-brush-pen-red-ink-brush-ink-leave-the-material-text.png"

    hide_message(image_path, secret_message, output_path)

    # Verify the hidden message
    extracted_message = reveal_message(output_path)
    print("Original Message:", secret_message)
    print("Extracted Message:", extracted_message)
```
